```
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import random

x = np.arange(1,61,1)
performance = np.random.randint(0,30, 60)
mean=[np.mean(performance)]*len(x)
median=[np.median(performance)]*len(x)
mode=[stats.mode(performance).mode]*len(x)

fig, x_axis = plt.subplots()
x_axis.plot(x, performance, color="green",label='Marks', marker='p')
mean_line = x_axis.plot(x,mean,color="red" ,label='Mean', linestyle='dotted')
median_line = x_axis.plot(x,median, color="blue",label='Median', linestyle='--')
mode_line = x_axis.plot(x,mode, color="blue",label='Mode', linestyle='-')
x_axis.legend(fancybox=True, framealpha=0.5,loc="best")
plt.ylabel('Marks')
plt.xlabel('Roll')
plt.title("Student's Marks")
plt.show()

```
